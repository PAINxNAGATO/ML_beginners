```python
df['mean score'] = ((df['math score'] + df['reading score'] + df['writing score'])/3).round()
df.head(20)
```

**LABEL ENCODING**
label encoding is an methode to convert categorial data into numerical data by mapping each unique value to a unique number, in the order to their apperance.

```python
from sklearn.preprocessing import LabelEncoder
# calling an LabelEncoder class form the sk learn module.
lc = LabelEncoder()
# created an instance of an class "LabelEncoder" on the varible lc. 
df['gender'] = lc.fit_transform(df['gender'])
# this function .fit_transform is label encoding gender data frame and then we reassign it tp gender.
df['race/ethnicity'] = lc.fit_transform(df['race/ethnicity'])
df['parental level of education'] = lc.fit_transform(df['parental level of education'])
df['lunch'] = lc.fit_transform(df['lunch'])
df['test preparation course'] = lc.fit_transform(df['test preparation course'])
df.head(20)
```

```python
df = df.drop(['math score', 'writing score', 'reading score'],axis = 1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
y = df['mean score']
x = df.drop(['mean score'], axis  = 1)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
# sk learn is an python module which has an class LogisticsRegression whose instance can be called to train the data .
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = LogisticRegression(solver = 'liblinear', random_state = 0)
# its an logistics regression model imported from sklearn using the solver as - liblinear .
# its an good solver for small data set .
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
difference = abs(predictions - y_test)
```
